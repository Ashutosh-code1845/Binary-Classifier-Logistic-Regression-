```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np

iris = datasets.load_iris()

# print(iris.DESCR)

# using a single feature so as to plot the sigmoid function
#to consider all 4 fetaures we will simply change syntax to
# x = iris.data
x = iris.data[:, 3:]

# 2 corressponds to flower virginica
y = (iris.target == 2).astype(int)

clf = LogisticRegression()
clf.fit(x,y)

prediction = clf.predict([[2.6]])
print(prediction)

# Plotting shifted sigmoid function
x_axis = np.linspace(0,3,1000).reshape(-1,1)
y_axis = clf.predict_proba(x_axis)
plt.plot(x_axis,y_axis[:,1])
plt.xlabel("Petal Length")
plt.ylabel("Probability")
plt.show()
```
